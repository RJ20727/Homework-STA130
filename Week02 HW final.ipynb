```python
#4. Provided by Chatbot ---link:https://chatgpt.com/share/66eaee52-29fc-8012-b01a-bc524f61f287
import random
import re
from collections import defaultdict

class MarkovChatBot:
    def __init__(self):
        # This will store word pairs in the form of {current_word: [list_of_next_words]}
        self.markov_chain = defaultdict(list)

    def train(self, text):
        # Clean and tokenize the input text
        words = re.findall(r'\b\w+\b', text.lower())
        
        # Build the Markov chain
        for i in range(len(words) - 1):# why len(words) - 1. explanation：it ensure the input of consecutive words to avoid the situation which is out of index
            current_word = words[i]
            next_word = words[i + 1]
            self.markov_chain[current_word].append(next_word)
    
    def generate_sentence(self, length=15):
        # Randomly choose a starting word
        current_word = random.choice(list(self.markov_chain.keys()))   #This loop is used to generate a sequence of words to build the response sentence.
                                                                       #length - 1 is used because we already have one starting word (current_word). We need to generate the remaining words to reach the desired total length.
        sentence = [current_word]# a new list
        
        for _ in range(length - 1):
            next_words = self.markov_chain.get(current_word, None)
            if not next_words:
                break  # If there are no next words, terminate the sentence
            next_word = random.choice(next_words)
            sentence.append(next_word)
            current_word = next_word
        
        return ' '.join(sentence) #This method ensures that the generated sentence follows the patterns learned from the input text, while also handling cases where no further words can be predicted.

    def chat(self, input_text):
        # Train on the input text to adjust the model
        self.train(input_text)
        
        # Generate a response
        response = self.generate_sentence()
        return response


# Sample Usage
if __name__ == "__main__":
    # Create an instance of the chatbot
    bot = MarkovChatBot()
    
    # Train the bot with some sample text (you can replace this with any text)
    sample_text = """
    Hello, how are you today? I hope you're having a great day!
    This chatbot uses a Markov Chain to generate responses based on the input text.
    Feel free to ask anything, and I'll try to respond in kind.
    """
    
    bot.train(sample_text)
    
    # Chat with the bot to generate the result
    user_input = "Tell me something interesting."
    print(f"User: {user_input}")
    response = bot.chat(user_input)
    print(f"Bot: {response}")
    

```
